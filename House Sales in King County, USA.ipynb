# House Sales in King County, USA

This dataset contains house sale prices for King County, which includes Seattle. It includes homes sold between May 2014 and May 2015

__id__: A notation for a house


__date__: Date house was sold


__price__: Price is prediction target


__bedrooms__: Number of bedrooms


__bathrooms__: Number of bathrooms


__sqft_living__: Square footage of the home


__sqft_lot__: Square footage of the lot


__floors__: Total floors (levels) in house


__waterfront__: House which has a view to a waterfront


__view__: Has been viewed


__condition__: Has been viewed


__grade__: Has been viewed


__sqft_above__: Square footage of house apart from basement


__sqft_basement__: Square footage of the basement


__yr_built__: Built Year


__yr_renovated__: Year when house was renovated


__zipcode__: Zip code


__lat__: Latitude coordinate


__long__: Longitude coordinate


__sqft_living15__: Living room area in 2015(implies-- some renovations) This might or might not have affected the lotsize area


__sqft_lot15__: LotSize area in 2015(implies-- some renovations)



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

## Importing Data Sets 

```python
file_name="https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DA0101EN/coursera/project/kc_house_data_NaN.csv"
df=pd.read_csv(file_name)
```

## Data Wrangling 

```python
df.drop(["id", "Unnamed: 0"], axis=1, inplace= True)   #using pandas "drop" method to drop columns "id" & "Unnamed: 0" 
df.describe()                    #using pandas "describe" method to get the statistical summary of the dataframe after dropping "id" & "Unnamed: 0" columns from the dataframe
```

```python
print("the number of NaN values for the column bedrooms :", df["bedrooms"].isnull().sum())
print("the number of NaN values for the column bathrooms :", df["bathrooms"].isnull().sum())
```

```python
#since the data type of columns "bedrooms" & "bathrooms" is of float type so we can replace the missing values with columns mean
df["bedrooms"].replace(np.nan, df["bedrooms"].mean(), inplace= True)
df["bathrooms"].replace(np.nan, df["bathrooms"].mean(), inplace= True)
```

```python
#to verify the changes made above in columns "bedrooms" & "bathrooms"
print("the no. of NaN values for the column bedrooms : ", df["bedrooms"].isnull().sum())
print("the no. of NaN values for the column bathrooms :", df["bathrooms"].isnull().sum())
```

## Model Development 

```python
#Fitting a linear regression model using "long"(longitude) feature 
x = df[["long"]]
y = df["price"]
lm = LinearRegression()
lm.fit(x,y)

#Calculating R^2
lm.score(x,y)
print("The calculated R-squared is :", lm.score(x,y))
```

```python
#Fitting a linear regression model using "sqft_living" feature
X = df[["sqft_living"]]
Y= df["price"]
lm= LinearRegression()
lm.fit(X,Y)
#Calculating the R-squared
lm.score(X,Y)
print("The R^2 is :", lm.score(X,Y))
```

```python
#Fitting a linear regression model to predict the 'price' using the list of features and calculating the R-squared

features = ["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]
Z = df[features]
Y = df["price"]
lm= LinearRegression()
lm.fit(Z, Y)

lm.score(Z, Y)
print("The R^2 is :", lm.score(Z, Y))
```

```python
#Creating a pipeline object using the given list to predict the 'price', fitting the object using the features in the list "features", and calculating the R^2
Input=[('scale',StandardScaler()),('polynomial', PolynomialFeatures(include_bias=False)),('model',LinearRegression())]
pipe = Pipeline(Input)
pipe.fit(Z, Y)
pipe.score(Z, Y)
```
