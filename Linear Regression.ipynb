```python
#Importing the required libraries

import pandas as pd
import numpy as np
import numpy.random as rnd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import StandardScaler
```

# Part 1

```python
#Setting the random seed

np.random.seed(10)
```

```python
#Making 200 samples and storing them into "X" variable

X=np.random.rand(200)
```

```python
#Checking the minimum, maximum and length of the sample

min(X),max(X),len(X)
```

```python
#Creatig noise term and assigning it to y variable

noise=np.random.randn(200)
Y=(12*X)-4+noise
```

```python
#Checking minimum, maximum and length of the data

min(Y),max(Y),len(Y)
```

```python
#Using matplotlib to create a plot

plt.subplots(figsize=(8, 6))
plt.scatter(X,Y,c="green",s=20)
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plotting of data points")
```

```python
#Reshaping "X" and "Y" variable

X=X.reshape(-1,1)
X.shape
Y=Y.reshape(-1,1)
Y.shape
```

```python
#Splittingt the data into training and testing parts

X=X.reshape(-1,1)
Y=Y.reshape(-1,1)

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = .2, random_state=4)
len(X_train),len(X_test)
```

```python
#Creating linear regression object and predicting the y variable

lr=LinearRegression()
lr.fit(X_train,Y_train)
Y_pred=lr.predict(X_test)
```

```python
#Checking the accuracy and performance of the model

print('Method: {}'.format(lr))
print('\nIntercept: {:0.2f}'.format(float(lr.intercept_)))
coeff_table=pd.DataFrame(np.transpose(lr.coef_),columns=['Coefficients'])
print('\n')
print(coeff_table)
    
#Calculating Mean absolute Error, Mean squared error and root mean squared error

print('\nR2: {:0.2f}'.format(metrics.r2_score(Y_test, Y_pred)))
print('Mean Absolute Error: {:0.2f}'.format(metrics.mean_absolute_error(Y_test, Y_pred)))  
print('Mean Squared Error: {:0.2f}'.format(metrics.mean_squared_error(Y_test, Y_pred)))  
print('Root Mean Squared Error: {:0.2f}'.format(np.sqrt(metrics.mean_squared_error(Y_test, Y_pred))))
```

```python
#Using matplotlib to create the regression chart

plt.subplots(figsize=(8, 6))
plt.scatter(X_test,Y_test,c="green")
plt.plot(X_test,Y_pred,c="red")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.title("Plotting fitted regression line with the data points")
```

# Part 2

```python
#Setting random seed

np.random.seed(10)
x=np.random.randint(-300,300,200)
x_req=x/100
x_req
```

```python
#Checking the minimum, maximum and length 

min(x_req),max(x_req),len(x_req)
```

```python
#Creating linear regression equation

y=.75*(x_req)+2*((x_req**2))+1

min(y),len(y)
```

```python
#Using matplotlib to create the chart

plt.subplots(figsize=(8, 6))
plt.scatter(x_req,y,c="green",s=20)
plt.xlabel("X values")
plt.ylabel("y values")
plt.title("Plotting of data points")
```

```python
#Reshaping the data points

x_req=x_req.reshape(-1,1)
x_req.shape
y=y.reshape(-1,1)
y.shape
```

```python
#Creating training and testing samples

x_train, x_test, y_train, y_test = train_test_split(x_req, y, test_size = .2, random_state=4)
print("Training and Testing samples:", len(x_train), len(x_test))
```

```python
#Creating linear regression object and predicting y

lr=LinearRegression()
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
#Checking the accuracy and performance of the model

print('Method: {}'.format(lr))
print('\nIntercept: {:0.2f}'.format(float(lr.intercept_)))
coeff_table=pd.DataFrame(np.transpose(lr.coef_),columns=['Coefficients'])
print('\n')
print(coeff_table)
    
#MAE,MSE and RMSE

print('\nR2: {:0.2f}'.format(metrics.r2_score(y_test, y_pred)))
print('Mean Absolute Error: {:0.2f}'.format(metrics.mean_absolute_error(y_test, y_pred)))  
print('Mean Squared Error: {:0.2f}'.format(metrics.mean_squared_error(y_test, y_pred)))  
print('Root Mean Squared Error: {:0.2f}'.format(np.sqrt(metrics.mean_squared_error(y_test, y_pred))))
```

```python
#Using matplotlib to create the chart

plt.subplots(figsize=(8, 6))
plt.scatter(x_test,y_test,c="green")
plt.plot(x_test,y_pred,c="red")
plt.xlabel("X values")
plt.ylabel("y values")
plt.title("Plotting of data points")
```

```python
scale=StandardScaler()
x_req=scale.fit_transform(x_req)

poly=PolynomialFeatures(degree=2,interaction_only=False)

x=poly.fit_transform(x_req)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .2, random_state=42)

poly.fit(x_train,y_train)

model=LinearRegression()
model.fit(x_train,y_train)

y_pred=model.predict(x_test)
print('\nR2: {:0.2f}'.format(metrics.r2_score(y_test, y_pred)))
```

```python
#Creating samples

x_sample=np.linspace(-3,3,100).reshape(-1,1)
y_sample=.75*(x_sample)+2*(x_sample**2)+1
x_sample.shape,y_sample.shape
```

```python
#Creating samples for transforming and fitting the data

x_sample=scale.fit_transform(x_sample)

x_sample_poly=poly.fit_transform(x_sample)
x_sample_poly.shape
```

```python
y_pred_poly=model.predict(x_sample_poly)
y_pred_poly.shape, y_sample.shape
```

```python
print('\nR2: {:0.2f}'.format(metrics.r2_score(y_sample, y_pred_poly)))
```

```python
#Creating the chart

plt.subplots(figsize=(8, 6))
plt.scatter(x_sample,y_sample,c="Green")
plt.plot(x_sample,y_pred_poly,c="Red")
plt.xlabel("X values")
plt.ylabel("y values")
plt.title("Plotting data points on the fitted polynomial line")
```
