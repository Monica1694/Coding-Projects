# Axes labels, legend, grid 

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")              #To label the x-axis
plt.ylabel("Temperature")       #To label the y-axis
plt.title("Weather")            #To set the title

plt.plot(days, max_t)
plt.plot(days, min_t)
plt.plot(days, avg_t)
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

#To set the legend, first you need to set label in each of the lines
plt.plot(days, max_t, label= "Max")        #Setting label to the 1st line
plt.plot(days, min_t, label = "Min")       #Setting label to the 2nd line
plt.plot(days, avg_t, label = "Avg")       #Setting label to the 3rd line

#Setting legend property
plt.legend()
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")

#To set the legend, first you need to set label in each of the lines
plt.plot(days, max_t, label= "Max")        #Setting label to the 1st line
plt.plot(days, min_t, label = "Min")       #Setting label to the 2nd line
plt.plot(days, avg_t, label = "Avg")       #Setting label to the 3rd line

#Setting legend property
plt.legend(loc = 8)              #Setting location "loc" property to "8" or "lower center" to place the legend in the lower center       
```

# Bar Chart

### Simple Bar chart showing revenues of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
```

```python
import numpy as np
```

```python
#creating array of company list to make it apt for "bar" function by using "np.arange()"

xpos = np.arange(len(company))      #xpos if for "x position"
xpos
```

```python
#Using the above array we can now plot the bar chart for the above company and revenue data

plt.bar(xpos, revenue)
```

```python
plt.xticks(xpos, company)  #"xticks" method will replace the array elements (xpos) with the company labels
plt.bar(xpos, revenue)
```

```python
#Since we already have company name so we need to label just y-axis and set title

xpos = np.arange(len(company))
plt.xticks(xpos, company)
plt.ylabel("revenue(in US billions)")  #Labelling "y-axis" with "revenue in US billion dollars"
plt.title("US Tech Stocks")            #Setting title "US Tech Stocks"
plt.bar(xpos, revenue)
```

```python
#we can pass the "label" parameter in "bar function" that will be displayed in the "legend"

xpos = np.arange(len(company))
plt.xticks(xpos, company)
plt.ylabel("revenue(in US billions)")  
plt.title("US Tech Stocks")            
plt.bar(xpos, revenue, label = "Revenue") #"label" parameter will be displayed in the legend
plt.legend()
```

### Multiple Bars showing revenue and profit of major US tech companies

```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]
```

```python
xpos = np.arange(len(company))
plt.xticks(xpos, company)
plt.ylabel("revenue(in US billions)")  
plt.title("US Tech Stocks")            
plt.bar(xpos, revenue, label = "Revenue") 
plt.bar(xpos, profit, label = "Profit")   #here bar function has "xpos" & "profit"
plt.legend()
```

```python
xpos = np.arange(len(company))
xpos
```

```python
#Creating side by side bars instead of single bar displaying two values

plt.xticks(xpos, company)
plt.ylabel("revenue(in US billions)")  
plt.title("US Tech Stocks")     

#we are subtracting 0.2 from xpos so that 0.2 will be subtracted from each of the elements of the array, e.g, 0-0.2, 1-0.2, & so on
plt.bar(xpos-0.2, revenue, label = "Revenue")  

#we are adding 0.2 to xpos so that 0.2 will be added to each of the elements of the array, doing these steps will create 2 bars
plt.bar(xpos+0.2, profit, label = "Profit")   
plt.legend()
```

```python
#Adjusting the width of the bars by passing "width" parameter in bar function
plt.xticks(xpos, company)
plt.ylabel("revenue(in US billions)")  
plt.title("US Tech Stocks")     
plt.bar(xpos-0.2, revenue, width= 0.4, label = "Revenue")  #"width" parameter to adjust the width of the bars  
plt.bar(xpos+0.2, profit, width= 0.4, label = "Profit")   #"width" parameter to adjust the width of the bars
plt.legend()
```

```python
#To create horizontal bars instead of vertical bars
plt.yticks(xpos, company)
plt.xlabel("revenue(in US billions)")  
plt.title("US Tech Stocks")     
plt.barh(xpos-0.2, revenue, label = "Revenue")    
plt.barh(xpos+0.2, profit, label = "Profit")   
plt.legend()
```

# Histograms 

### In histograms, x-axis contains a variable and y-axis will be a frequency of that variable

#### We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic 

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]

#For plotting histogram, use the "hist" function & simply pass in the array or the list

plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins = 3)   #Passing "bins" parameter to create desired no. of bins
```

```python
plt.hist(blood_sugar, bins = 3, rwidth = 0.95) #passing "rwidth" parameter to separate bins
```

```python
plt.hist(blood_sugar, bins =[80, 100, 125, 150], rwidth = 0.95)  #We have specified bins in the bins parameter

#80 to 100 normal range, 100 to 125 is pre-diabetic and anything beyond 125 is diabetic so we roughly set the max limit to 150
```

```python
plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = 0.95, color = "g") #passing color parameter
```

```python
#Passing "histtype" parameter and setting it to "step"
plt.hist(blood_sugar, bins = [80,100,125,150], rwidth = 0.95, color= "g", histtype = "step")
```

### Multiple data samples in a histogram 

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
#For plotting the data side by side we need to create a list or an array of two lists

plt.hist([blood_sugar_men, blood_sugar_women], bins = [80,100,125,150], rwidth = 0.95, color=["green", "orange"])  
```

```python
#For labelling bars we need to pass "label" parameter in hist function

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=["green", "orange"], 
         label=["men", "women"])

plt.legend()
```

```python
#Labelling "x-axis", "y-axis" and setting the title

plt.xlabel("Sugar range")
plt.ylabel("Total no. of patients")
plt.title("Blood sugar analysis")

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=["green", "orange"], 
         label=["men", "women"])
plt.legend()
```

```python
#Passing "orientatiion" parameter in "hist" function to create horizontal bars
plt.xlabel("Total no. of patients")
plt.ylabel("Sugar range")
plt.title("Blood sugar analysis")

plt.hist([blood_sugar_men, blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=["green", "orange"], 
         label=["men", "women"], orientation="horizontal")
plt.legend()
```

# Pie Chart

### Draw Pie Chart to track down Home Expenses 

```python
exp_vals = [1400,600,300,410,250]                                                 #Expense values
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]  #Expense labels
```

```python
#To plot a pie chart use "pie" function 

plt.pie(exp_vals, labels = exp_labels)   #Passing "labels" parameter to label each wedge
```

```python
#For perfect shape of the pie chart pass "plt.axis" to be equal

plt.axis = "equal"
plt.pie(exp_vals, labels = exp_labels)
```

```python
#To get rid of the extra information in the chart pass the "plt.show()" function

plt.axis = "equal"
plt.pie(exp_vals, labels = exp_labels)
plt.show()       #Passing the "show" function to get rid of extra text in the pie chart
```

```python
#Passing "radius" parameter in pie function to increase the size of the pie chart

plt.axis ="equal"
plt.pie(exp_vals, labels = exp_labels, radius = 1.5)
plt.show()
```

### Show percentages for every pie. Specify radius to increase chart size 

```python
#To show percentages we need to pass "autopct" parameter in pie function, where it has a string "%f%%"
#You need to put desired no. format before "f" in the given string ("%f%%")
#To get percentages with 2 decimal points, write "%0.2f%%"

plt.axis = "equal"
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.2f%%")
plt.show()
```

```python
#To get 1 decimal point, write "%0.1f%%"

plt.axis = "equal"
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.1f%%")
plt.show()
```

```python
#To get no decimal point(s) and just the plain percentage, write "%0.0f%%"

plt.axis = "equal"
plt.pie(exp_vals, labels = exp_labels, radius = 1.5, autopct = "%0.0f%%")
plt.show()
```

```python
#Passing "shadow" parameter
plt.pie(exp_vals, labels = exp_labels, autopct = "%0.1f%%", radius = 1.5, shadow = True)
plt.show()
```

### Explode 

```python
#Taking out the pies (wedges) from the pie by passing "explode" parameter, and in the explode list 5 values will go in as 
#there are 5 pies in the pie

plt.pie(exp_vals, labels = exp_labels, autopct = "%0.1f%%", radius = 1.5, shadow = True, explode = [0,0,0,0,0])
plt.show()
```

```python
#To take out "food" pie we need to change the 2nd value in the "explode" parameter "list"

plt.pie(exp_vals, labels = exp_labels, autopct = "%0.1f%%", radius = 1.5, shadow = True, explode = [0,0.2,0,0,0])
plt.show()
```

```python
#Taking out "food" & "car" pies

plt.pie(exp_vals, labels = exp_labels, autopct = "%0.1f%%", radius = 1.5, shadow = True, explode = [0, 0.5, 0, 0.5, 0])
plt.show()
```

### Angle properties 

```python
#To start pie from different angle, use "startangle" parameter and specify the desired "angle degree"
#Here we want to start pie from 45 degree angle

plt.pie(exp_vals, labels = exp_labels, autopct = "%0.1f%%", radius = 1.5, startangle = 45)
plt.show()
```
